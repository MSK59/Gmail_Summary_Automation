```python
from Gmail_Handler import GmailHandler
from LLM_Processor import LLM_Processor
from googleapiclient.discovery import build
import pandas as pd
import time

def main():
    # Initialize Gmail handler and authenticate
    gmail_handler = GmailHandler()  
    start_time = time.asctime(time.localtime())
    creds = gmail_handler.authenticate_gmail()
    
    # Build Gmail API service
    service = build('gmail', 'v1', credentials=creds)
    
    # Fetch unread emails with full bodies
    emails = gmail_handler.fetch_emails_full_body(service)
    
    # Initialize the LLM processor
    summarizer = LLM_Processor()
        
    # Process emails through the LLM
    #results = summarizer.process_emails(emails)
    results = summarizer.process_emails_in_parallel(emails, max_concurrent=5)
    
    # Display results in console
    if results:
        summarizer.display_results(results)
        
        # Add email links to the results
        for i in range(min(len(results), len(emails))):
            results[i]['link'] = emails[i]['link']
        
        # Create DataFrame and sort by importance
        results_df = pd.DataFrame.from_dict(results)
        results_df.sort_values(by='importance_score', ascending=False, inplace=True)
        
        # Save results to CSV
        results_df.to_csv('email_summaries.csv', index=False)
        print("\nResults saved to 'email_summaries.csv'.")
    else:
        print("No emails were successfully processed.")
    end_time = time.asctime(time.localtime())
    print(start_time, end_time, sep = '\n')
if __name__ == "__main__":
    main()
```
